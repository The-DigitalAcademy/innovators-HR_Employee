```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv("HR-Employee-Attrition.csv")
df
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
#Transforming Attrition to numeric values
df['Attrition'] = np.where(df['Attrition']=='Yes',1,0)
```

```python
df.hist(bins=15,figsize=(10,8), edgecolor="black")
plt.tight_layout()
```

```python
f,ax = plt.subplots(figsize=(10,10))
sns.heatmap(df.corr(), annot=True, cmap ="YlGnBu", linewidths=.5, fmt= '.1f',ax=ax)
```

```python
values= df['BusinessTravel'].value_counts()
labels= df['BusinessTravel'].unique().tolist()

plt.pie(values, labels=labels, radius =1,autopct='%1.1f%%')

plt.show()
print(values)
```
